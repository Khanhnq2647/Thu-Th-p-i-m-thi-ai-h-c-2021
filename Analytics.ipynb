```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import ast
```

```python
df = pd.read_csv("Diemthi2021.csv")
df.head()
```

```python
df= df.rename(columns={"Cum_thi":"Ma_cum_thi"})
```

```python
with open('matinh.txt',mode = 'r', encoding = 'utf8') as file:
    data = file.read().replace("\n","")
```

```python
mapper = ast.literal_eval(data)
```

```python
df['Cum_thi'] = df['Ma_cum_thi'].map(mapper)
```

```python
df.head()
```

```python
Toan = df.Toan.value_counts().sort_index()
```

```python
fig, ax = plt.subplots(figsize = (16,8))
sns.barplot(data= df, y = Toan.values, x = Toan.index,palette="Blues_d")
plt.grid(c='white', linewidth=2, fillstyle='full', b=True)
```
